## 2. Data Loading

We will use Yahoo Finance data through the yfinance library.

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

print('Libraries imported successfully!')
```

## 3. Data Collection

Let's download stock data for Apple (AAPL) as an example.

```python
# Define parameters
ticker = 'AAPL'
end_date = datetime.now()
start_date = end_date - timedelta(days=365)

# Download data
print(f'Downloading data for {ticker}...')
stock = yf.Ticker(ticker)
df = stock.history(start=start_date, end=end_date + timedelta(days=1))

print(f'Data downloaded: {len(df)} trading days')
print(f'Date range: {df.index[0].strftime("%Y-%m-%d")} to {df.index[-1].strftime("%Y-%m-%d")}')
```

## 4. Data Exploration

```python
# View first few rows
df.head()
```

```python
# Check data types and missing values
print('Data types:')
print(df.dtypes)
print('\nMissing values:')
print(df.isnull().sum())
```

## 5. Data Cleaning

The data from Yahoo Finance is generally clean, but we should verify.

```python
# Check for any missing values
if df.isnull().sum().sum() == 0:
    print('No missing values found. Data is clean!')
else:
    print('Missing values found. Cleaning data...')
    df = df.dropna()
    print('Missing values removed.')
```

## 6. Analysis

```python
# Calculate daily returns
df['Daily_Return'] = df['Close'].pct_change()

# Calculate moving averages
df['MA_20'] = df['Close'].rolling(window=20).mean()
df['MA_50'] = df['Close'].rolling(window=50).mean()

# Display updated dataframe
df.tail(10)
```

```python
# Summary statistics
print('Summary Statistics for', ticker)
print('='*50)
print(f'Starting Price: ${df["Close"].iloc[0]:.2f}')
print(f'Ending Price: ${df["Close"].iloc[-1]:.2f}')
print(f'Total Return: {((df["Close"].iloc[-1]/df["Close"].iloc[0])-1)*100:.2f}%')
print(f'Average Daily Return: {df["Daily_Return"].mean()*100:.2f}%')
print(f'Daily Return Std Dev: {df["Daily_Return"].std()*100:.2f}%')
```

## 7. Visualization

```python
# Create price chart
fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df.index, df['Close'], label='Close Price', linewidth=2)
ax.plot(df.index, df['MA_20'], label='20-Day MA', alpha=0.7)
ax.plot(df.index, df['MA_50'], label='50-Day MA', alpha=0.7)
ax.set_title(f'{ticker} Stock Price Analysis', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.legend()
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Create volume chart
fig, ax = plt.subplots(figsize=(12, 4))
ax.bar(df.index, df['Volume'], color='steelblue', alpha=0.7)
ax.set_title(f'{ticker} Trading Volume', fontsize=14)
ax.set_xlabel('Date')
ax.set_ylabel('Volume')
ax.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
